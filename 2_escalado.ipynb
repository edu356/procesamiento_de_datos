# Escalado
Conjunto de datos: iris (incluido en sklearn)
Este conjunto de datos es compacto y perfecto para demostrar técnicas de escalado.

```python
# Importar librerías necesarias
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos
iris = load_iris(as_frame=True)
data = iris.data
```

```python
# Visualizar las estadísticas iniciales
print("Estadísticas iniciales:")
data.describe()
```

```python
# Aplicar MinMaxScaler
minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(minmax_scaler.fit_transform(data), columns=data.columns)
```

```python
# Aplicar StandardScaler
standard_scaler = StandardScaler()
data_standard = pd.DataFrame(standard_scaler.fit_transform(data), columns=data.columns)
```

4. ¿Qué técnica de escalado crees que es más adecuada para este conjunto de datos? ¿Por qué?

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Cargar el dataset Iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Escaladores
scaler_minmax = MinMaxScaler()
scaler_standard = StandardScaler()

# Aplicar escalado
data_minmax = pd.DataFrame(scaler_minmax.fit_transform(data), columns=data.columns)
data_standard = pd.DataFrame(scaler_standard.fit_transform(data), columns=data.columns)

# Visualizar comparaciones
feature = 'sepal length (cm)'
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

axs[0].hist(data[feature], bins=20, alpha=0.6, color='blue')
axs[0].set_title(f'Original - {feature}')
axs[0].set_xlabel(feature)
axs[0].set_ylabel('Frecuencia')

axs[1].hist(data_minmax[feature], bins=20, alpha=0.6, color='green')
axs[1].set_title(f'MinMaxScaler - {feature}')
axs[1].set_xlabel('Valor Escalado')
axs[1].set_ylabel('Frecuencia')

axs[2].hist(data_standard[feature], bins=20, alpha=0.6, color='red')
axs[2].set_title(f'StandardScaler - {feature}')
axs[2].set_xlabel('Valor Estandarizado')
axs[2].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()

# Recomendación basada en distribución
if abs(data_standard[feature].skew()) < abs(data_minmax[feature].skew()):
    print("StandardScaler conserva mejor la simetría de la distribución. Es más adecuado para modelos lineales o PCA.")
else:
    print("MinMaxScaler conserva mejor la forma original. Es útil para modelos basados en distancia o redes neuronales.")
```
